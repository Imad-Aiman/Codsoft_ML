Importing the dependencies

```python
import pandas as pd
import matplotlib as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading of the dataset to a pandas dataframe
ccdata=pd.read_csv('/content/creditcard.csv')
#reading the first 5 rows of the dataset
ccdata.head()
```

```python
#printing information about the data
ccdata.info()
```

```python
#checking the no.of null values in each column
ccdata.isnull().sum()
```

```python
#distribution of legit transaction & fradulent transaction
ccdata['Class'].value_counts()
```

1. Above, 0 represents normal transaction
   and 1 represents fraudelent transaction.
2. By looking at the count values of 0 and 1
   above we can clearly identify that the data is overfitting which may result in biased outputs/predictions.

```python
#sperating the data for analysis
legit=ccdata[ccdata.Class == 0]
fraud=ccdata[ccdata.Class == 1]
print(legit.head())
print(fraud.head())
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
print(legit.Amount.describe())
print(fraud.Amount.describe( ))
```

```python
#compare the values for both transactions
ccdata.groupby('Class').mean()
```

Under-Sampling
 Building a sample dataset containing similar distribution of legit and fraudental transactions

```python
#creating a sample dataset with 492 random legit transactions from overall legit transactions
legit_sample=legit.sample(n=492)
```

Concatenating the two dataframes

```python
#joining the legit_sample and fraud to get a single dataframe
new=pd.concat([legit_sample,fraud],axis=0)
```

```python
#checking the first 5 rows of new dataset
new.head()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

Splitting the Data into Training and Testing Data

```python
X=new.drop(columns='Class',axis=1)
Y=new['Class']
print(X)
print(Y)
```

Split the data into Traning and Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Traning[Logistic Regression]

```python
m=LogisticRegression()
#traning the LogisticRegression model with traning data
m.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on traning data
X_train_prediction=m.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy on traning data:',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=m.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy on test data:',test_data_accuracy)
```
